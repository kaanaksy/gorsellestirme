```python
import pandas as pd;
import matplotlib.pyplot as plt;
import numpy as np;
```

```python
%matplotlib inline
plt.style.use("fivethirtyeight")
```

```python
# ALAN GRAFİĞİ
# Alan grafiği için sütunlardaki veriler ya negatif ya da pozitif olmalı

df = pd.DataFrame(abs(np.random.randn(10,4)), columns=list("ABCD"))
df.head()
```

```python
df["A"].plot.area()
```

```python
df.plot.area()
```

```python
# Üst üste bindirilmiş alan grafikleri için stacked=False kullanılır

df.plot.area(stacked=False)
```

# NOT: Alan grafiği oluştururken kullanılan veri setinde eksik veri varsa bu veri =0 kabul edilir

```python
df.plot.scatter(x="A", y="B")
```

```python
iris = pd.read_csv("iris.txt", sep=",")
```

```python
iris.head()
```

```python
iris.plot.scatter(x="Canak_yaprak_boyu", y="Canak_yaprak_eni")
```

```python
ax = iris.plot.scatter(x="Canak_yaprak_boyu", y="Canak_yaprak_eni", color="Blue", label="Canak")
iris.plot.scatter(x="Tac_yaprak_boyu", y="Tac_yaprak_eni", color="Red", ax=ax)
```

```python
# İki değişkeni karşılaştırırken her bir noktanın rengini ayarlamak için c argümanı eklenir
#s=100 değeri noktanın boyutunu belirler

iris.plot.scatter(x="Canak_yaprak_boyu", y="Canak_yaprak_eni", c="Tac_yaprak_boyu", s=100)
```

```python
iris.plot.scatter(x="Canak_yaprak_boyu", y="Canak_yaprak_eni", c="Tac_yaprak_boyu", s=iris["Tac_yaprak_boyu"]*50)
```

```python
#ALTIGEN (HEXAGONAL) GRAFİKLER
#Daha büyük daha gözlemli veriler ile çalışırken kullanılabilir
#gridsize grafikteki altıgenlerin boyutunu belirler. Değer azaldıkça boyut artar
df = pd.DataFrame({"a":np.random.randn(1000),
                  "b":np.random.randn(1000)})
df.plot.hexbin(x="a", y="b", gridsize=10)
```

```python
# PASTA GRAFİĞİ

s = pd.Series(3*abs(np.random.randn(5)), index=list("ABCDE"), name="Series")
s.plot.pie()
```

```python
df = pd.DataFrame(3*abs(np.random.randn(4,2)), index=list("abcd"), columns=list("xy"))
```

```python
#DataFrame veri setinin pasta grafiği için ya spesifik Y değeri girilmesi gerekir ya da subplot() metodyla oluşturulması gerekir.
df.plot.pie(subplots=True)
```

```python
s.plot.pie()
```
